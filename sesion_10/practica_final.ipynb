```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


class Calculadora:
    """
        Esto es una calculadora que nos permitirá realizar diferentes 
        operaciones
    """
    def __init__(self, memoria, pantalla, color):
        """
            Este es el constructos
        """
        self.memoria = memoria
        self.pantalla = pantalla
        self.color = color
        self.prohibido = 0
        plt.style.use("ggplot")

    def suma(self, x, y):
        return x + y

    def producto(self, x, y):
        return x*y

    def division(self, x, y):
        if y != self.prohibido:
            return x/y
        else:
            print("No es válida la división por cero en números de punto flotante... ")

    def factorial(self, n):
        if n > 0:
            if n == 1:
                return 1
            else:
                return n * self.factorial(n-1)
        else:
            raise ValueError("El factorial no está definido para números negativos")

    def fibonacci(self, n):
        if n > 0:
            fibo = [1, 1]
            
            for i in range(2, n):
                fib = fibo[i-1] + fibo[i-2]
                fibo.append(fib)
                
            return fibo[n-1]
        else:
            raise ValueError("Las sucesiones solo están definidas en los números naturales")

    def graficar(self, function, dominio, title, label, color):
        if self.pantalla == False:
            print("Lo siento no tengo la tecnología para graficar")
        else:
            imagen = [function(x) for x in dominio]
            
            plt.figure(figsize=(16, 10), dpi=180)
            plt.plot(dominio, imagen, label=label, color=color)
            plt.title(title)
            plt.legend()
            plt.show()
    
    def vp_anualidad(self, interest_rate, period_number=10):
        """
            Valor presente de una anualidad vencida fija.
        """
        present_value = 1/ ( 1 +  interest_rate) 
        return (1 + present_value**period_number)/interest_rate

    
```

```python
calculadora = Calculadora(
    256,
    True,
    "blue"
)
```

```python
calculadora.graficar(
    np.exp, range(1, 100), "Función Exponencial", "Exponencial", "darkblue"
)
```

```python
calculadora.graficar(
    np.sin,
    np.linspace(-10, 10, 100000),
    title="Función Seno",
    label="seno",
    color="orange"
)
calculadora.graficar(
    np.cos,
    np.linspace(-10, 10, 100000),
    title="Función Coseno",
    label="seno",
    color="orange"
)
```

```python
calculadora.graficar(
    calculadora.vp_anualidad,
    np.linspace(0, 1, 1000),
    title="Valor presente de una anualidad como función del interés",
    label="Valor Presente",
    color="orange"
)
```
