```python
### Tiene complejidad lineal
def factorial(x : int) -> int:
    producto = 1
    
    for n in range(1, x + 1):
        producto = producto * n
    
    return producto
```

```python
factorial(3), factorial(5), factorial(7), factorial(1)
```

## Definiciones Recursivas

```python
### Tiene complejidad lineal
def factorial_rec(x: int) -> int:
    if x == 1:
        return 1
    else:
        return x * factorial_rec( x - 1)
```

```python
factorial_rec(3), factorial_rec(5), factorial_rec(7), factorial_rec(1)
```

## Sucesi√≥n de Fibonacci

$$ a_1 = 1 , \:\: a_2 = 1 $$

$$ a_n = a_{n-1} + a_{n-2} $$

```python
def fibonacci(x):
    fib = [1, 1]

    for i in range(2, x):        
        fib.append(fib[i-1] + fib[i-2])
    
    return fib[x-1]
```

```python
fibonacci(4), fibonacci(5), fibonacci(10), fibonacci(100)
```

```python
def fibonacci(x):
    if x == 1 or x == 2:
        return 1
    else:
        return fibonacci(x-1) + fibonacci(x-2)
```

```python
fibonacci(4), fibonacci(5), fibonacci(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### List Comprehension

dominio = [x for x in range(1, 10)]
fact = [factorial(x) for x in dominio]
exp = [np.exp(x) for x in dominio]
log = [np.log(x) for x in dominio]
lin = [x for x in dominio]
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(8, 4), dpi=160)
plt.plot(dominio, fact, label="Factorial")
plt.plot(dominio, exp, label="Exponencial")
plt.plot(dominio, log, label="Logaritmo")
plt.plot(dominio, lin, label="Identidad")

plt.title("Comparaci√≥n de funciones")
plt.legend()
plt.show()
```

```python
dominio = [x for x in range(1, 101)]
sin = [np.sin(x) for x in dominio]
cos = [np.cos(x) for x in dominio]
tan = [np.tan(x) for x in dominio]
sec = [1/np.tan(x) for x in dominio]
```

```python
def function_plotter(function, domain, title, color):
    image = [function(x) for x in domain]
    
    plt.figure(figsize=(10, 8), dpi=120)
    plt.plot(domain, image, label=str(function), color=color)
    plt.title(title)
    plt.legend()
    plt.show()
    
```

```python
function_plotter(
    np.sin, 
    np.linspace(-10, 10, 10000), 
    title="Funcion Seno", 
    color="gold"
)
```

```python
function_plotter(np.cos, np.linspace(-10, 10, 10000), title="Funcion Seno")
```
