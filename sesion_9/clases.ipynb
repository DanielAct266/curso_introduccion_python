```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Persona:
    def __init__(self, edad: int, discapacitado: bool):
        """
            Constructor: Nos permite crear diferentes ejemplares,
            acá van los atributos
        """
        self.edad = edad
        self.discapacitado = discapacitado
    
    def correr(self):
        """
            Este es un método que indica correr
        """
        if self.discapacitado == True:
            print("Lo siento, pero no puedo correr")
        else:
            print("Wiii, estoy corriendo")
            
```

```python
### Por el momento seamos consistentes con los tipos de datos
pedrito = Persona(25, True)
```

```python
pedrito.correr()
```

```python
enrique = Persona(30, False)
```

```python
enrique.correr()
```

```python
class Calculadora:
    """
        Esto es una calculadora que nos permitirá realizar diferentes 
        operaciones
    """
    def __init__(self, memoria, pantalla, color):
        """
            Este es el constructos
        """
        self.memoria = memoria
        self.pantalla = pantalla
        self.color = color
        self.prohibido = 0
        plt.style.use("ggplot")
    
    def suma(self, x, y):
        return x + y
    
    def producto(self, x, y):
        return x*y
    
    def division(self, x, y):
        if y != self.prohibido:
            return x/y
        else:
            print("No es válida la división por cero")
    
    def factorial(self, n):
        if n == 1:
            return 1
        else:
            return n * self.factorial(n-1)
            
    def graficar(self, function, dominio):
        if self.pantalla == False:
            print("Lo siento no tengo la tecnología para graficar")
        else:
            imagen = [function(x) for x in dominio]
            plt.plot(dominio, imagen)
            plt.show()
        
```

```python
calculadora = Calculadora(10, True, "azul")
```

```python
calculadora.graficar(
    np.sin, 
    np.linspace(-10, 10, 1000)
)
```

```python
calculadora.graficar(
    calculadora.factorial,
    range(1, 10)
)
```
